# Insert one Document
Create a dummy index just to test one inserting 

```python
from pprint import pprint
from elasticsearch import Elasticsearch

es = Elasticsearch(
    "https://localhost:9200", 
    basic_auth=("elastic", "6AqhOxi*CPXYvCZl7Iln"), 
    verify_certs=False)
```

```python

es.indices.delete(index='my_index', ignore_unavailable=True)
es.indices.create(
    index="my_index",
    settings={
        "index": {
            "number_of_shards": 3,  # how many pieces the data is split into
            "number_of_replicas": 2  # how many copies of the data
        }
    },
)
```

```python
document = {
    'title': 'title', 
    'text': 'text', 
    'created_on': '2024-09-22',
}
response = es.index(index='my_index', body=document)
response
```

```python
print(response["result"])
```

```python
print(response["_id"])
```

```python
print(response["_index"])
```

# Insert multiple documents
just do the same step but in a for loop 

```python
import json 

dummy_data = json.load(open("./data/dummy_data.json"))
dummy_data
```

```python
def insert_document(document):
    response = es.index(index='my_index', body=document)
    return response

def print_info(response):
    print(f"""Document ID: {response['_id']} is '{response["result"]}'
    and is split into {response['_shards']['total']} shards.
""")
```

```python
for document in dummy_data: 
    response = insert_document(document)
    print_info(response)
```

# Manual mapping

```python
es.indices.delete(index="my_index", ignore_unavailable=True)
es.indices.create(index="my_index")

mapping = {
    'properties': {
        'created_on': {'type': 'date'},
        'text': {
            'type': 'text',
            '':  {
                'keyword': {
                    'type': 'keyword', 
                    'ignore_above': 256
                }
            }
        },
        'title': {
            'type': 'text', 
            'fields': {
                'keyword': {
                    'type': 'keyword', 
                    'ignore_above': 256
                }
            }
        }
    }
}
```

```python
es.indices.put_mapping(index='my_index', body=mapping)

index_mapping = es.indices.get_mapping(index='my_index')
pprint(index_mapping["my_index"]["mappings"]["properties"])
```
